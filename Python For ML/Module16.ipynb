```python
import numpy as numpy
import pandas as pd 
import matplotlib as plt
import seaborn as sns 
import plotly.express as px
```

```python
student = pd.read_csv('sns_data (1).csv',index_col='student_id')
student.head()
```

```python
sns.lineplot(data=student, x='week', y='attendance_rate', errorbar=None)
```

```python
#axes level
sns.lineplot(data=student, x='week', y='attendance_rate', errorbar=None, hue='gender')
```

```python
#figure level 
sns.relplot(kind='line',data= student, x='week', y='attendance_rate',hue='gender', errorbar=None)
```

```python
sns.relplot(kind='line',data= student, x='week', y='test_score',hue='class_level', errorbar=None)
```
