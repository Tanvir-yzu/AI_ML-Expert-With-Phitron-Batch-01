```python
#  Bar chart on real data 

df = pd.read_csv('student_data.csv') 

df
```

```python
status = df.groupby('CompletionStatus').size()
```

```python
plt.bar(status.index,status.values,color='skyblue',     # fill color
    edgecolor='black',   # border color
    linewidth=1.2 )
plt.xlabel('Completion Status') 
plt.ylabel('Count') 
plt.title('Bar Chart of Course completion')
plt.show()
```

```python

# Example bar chart with more style
plt.figure(figsize=(10,6))  # Set figure size

# Bar chart with color and edge
bars = plt.bar(
    status.index, 
    status.values, 
    color='skyblue',     # fill color
    edgecolor='black',   # border color
    linewidth=1.2        # border width
)

# Add grid
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Labels and title with font size
plt.xlabel('Completion Status', fontsize=12, fontweight='bold')
plt.ylabel('Count', fontsize=12, fontweight='bold')
plt.title('Bar Chart of Course Completion', fontsize=14, fontweight='bold')

# Rotate x-axis labels if long
plt.xticks(rotation=30, fontsize=10)

# Add value labels on top of bars
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 0.5, int(yval),
             ha='center', va='bottom', fontsize=10, fontweight='bold')

plt.tight_layout()  # Adjust layout to prevent clipping
plt.show()
```

```python
# location 

status = df.groupby('Location').size() 

plt.bar(status.index,status.values,edgecolor='black')

plt.xlabel('Location') 
plt.ylabel('Count') 
plt.title('Bar Chart of Location of Students')
plt.show()
```
