```python
mn = df['Total Mark'].min()
mx = df['Total Mark'].max()

df['Scaled Marks'] = df['Total Mark'].apply(lambda x : (x - mn)/ (mx - mn))
```

```python
df
```

```python
def grading_system(marks):
    if marks >= 260 :
        return "A+"
    elif marks >=250:
        return 'A'
    else :
        return 'A-'

df['grade'] = df['Total Mark'].apply(grading_system)
```

```python
df
```

```python
def marking_system(df): 
    a = df['Algorithm Marks'] * 2
    b = df['Data Structure Marks'] * 3
    c = df['Python Marks'] * 4
    return a+ b+ c
```

```python
df['Exceptional Marks'] = df.apply(marking_system, axis=1)
```
